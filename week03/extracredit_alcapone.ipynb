```python
# Importing Pandas and opening the capone.csv file for further manipulation. 

import pandas as pd

df = pd.read_csv('capone.csv')
df
```

```python
# This code parses out the numbers and the letter from the "number" column. This creates two new columns rather than overriding the original column. Note, we don't need the letter_parsed column, but I keep it for review. 
import re

def parse_number_letter(value):
    match = re.match(r"(\d+)([A-Za-z]+)?", str(value))
    if match:
        number = match.group(1)
        letter = match.group(2) if match.group(2) else None
        return pd.Series([number, letter])
    return pd.Series([None, None])

df[['number_parsed', 'letter_parsed']] = df['number'].apply(parse_number_letter)
df
```

```python
#Needed to use pip to instal num2words in the databricks env. 
%pip install num2words
import num2words

# Adding a new column that spells out the previously parsed number. 
def spell_number(value):
    try:
        return num2words.num2words(int(value))
    except (ValueError, TypeError):
        return None

df['number_spelled_out'] = df['number_parsed'].apply(spell_number)
df
```

```python
# Grab the length of the spelled out number. 
df['spelled_length'] = df['number_spelled_out'].apply(lambda x: len(x) if isinstance(x, str) else None)
df
```

```python
# Compare the length of the spelled out number to the response code and create a new column based on the comparison with True/False values. 
df['spelled_matches_response'] = df.apply(
    lambda row: row['spelled_length'] == row['response_code'], axis=1
)
df
```

```python
#veryfy the match column isn't assigning any false values. 
display(df['spelled_matches_response'].value_counts())
```
