```python
from pyspark.sql import functions as F


# Read the CSV
df = (
    spark.read
        .option("header", "true")
        .option("inferScheme", "true")
        .csv("/Workspace/Users/jcoffey@wustl.edu/data-5035-2026/week03/hospitals.csv")
)

# Aggregate by city
result = (
    df.groupBy("city")
        .agg(
            F.count("*").alias("num_hospitals"),
            F.sum("beds").alias("total_beds")
        )
)

result.show()
```
