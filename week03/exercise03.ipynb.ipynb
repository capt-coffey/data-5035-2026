```python
# Importing needed modules. 
import json 
import csv
```

```python
# Opening the imported json file. 
df = open('/Workspace/Users/jcoffey@wustl.edu/data-5035-2026/week03/negotiated_rates.json')

# loading the json file as a dictionary object. 
h = json.load(df)
```

```python
# Create the list we'll add to later. Initializing it here, so each time I run this code block, any existing data in the list is erased. 
out_of_network_list = []


# Creating the headers for the list / final csv file 
out_of_network_list.append(['name', 'billing_code_type', 'billing_code', 'description', 'service_code', 'billing_class', 'allowed_amount', 'billed_charged', 'npi'])

# Looping through the json file to fill each each row. I've used the nested loop to catch when there are multiple allowed_amounts in an out of network entry. 
for item in h['out_of_network']:
    for allowed in item['allowed_amounts']:
        row = [
            item.get('name'),
            item.get('billing_code_type'),
            item.get('billing_code'),
            item.get('description'),
            allowed.get('service_code'),
            allowed.get('billing_class'),
            allowed['payments']['allowed_amount'],
            allowed['payments']['providers'][0]['billed_charge'],
            allowed['payments']['providers'][0]['npi'][0]
        ]
        out_of_network_list.append(row)
        
#displaying
out_of_network_list
```

```python
# Creating the output csv file. 
with open('/Workspace/Users/jcoffey@wustl.edu/data-5035-2026/week03/out_of_network.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(out_of_network_list)
```
